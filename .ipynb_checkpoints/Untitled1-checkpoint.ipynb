```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#训练数据查看
sales = pd.read_csv('./train.csv')
sales.head()
```

```python
#商店数据查看
store = pd.read_csv('./store.csv')
rossm = pd.merge(sales,store,on='Store')
store.head()
```

```python
#商店销售数据与商店信息数据合并
rossm.head()
```

```python
#查看一周商店是否营业分布情况
fig, (axis1) = plt.subplots(1,1,figsize=(15,4))
sns.countplot(x='Open',hue='DayOfWeek',data=sales,ax = axis1)
```

```python
#查看四中商店类型的销售情况
sns.barplot(x="StoreType",y='Sales',data=rossm)
```

```python
#拆解字段
sales['Year'] = sales['Date'].apply(lambda x: int(str(x)[:4]))
sales['Month'] = sales['Date'].apply(lambda x:int(str(x)[5:7]))
sales['year_month'] =sales['Date'].apply(lambda x:str(x)[:7])
sales['Date'] = sales['Date'].apply(lambda x: (str(x)[:7]))
sales.head()
```

# 数据呈现周期性波动

```python
average_year_month = sales.groupby('year_month')['Sales'].mean()                #每月平均销售额
pct_change_year_month = sales.groupby('year_month')['Sales'].sum().pct_change() #每月销量波动情况
fig,(axis1,axis2) = plt.subplots(2,1,figsize=(15,8))
average_year_month.plot(legend=True,ax=axis1,marker='o')
pct_change_year_month.plot(legend=True,ax=axis2,marker='o')
```

```python
sns.countplot(x='Promo',data= sales)                                      #促销对比
fig,(axis1,axis2) = plt.subplots(1,2,figsize=(15,5))                      
sns.barplot(x='Promo',y='Sales',data=sales,ax= axis1)               #销售额与促销情况对比分析
sns.barplot(x='Promo',y='Customers',data=sales,ax= axis2)       #顾客数量与促销情况对比分析
```

```python
sns.countplot(x='Promo2',data= rossm)     #持续做促销和不促销对比分析
fig,(axis1,axis2) = plt.subplots(1,2,figsize=(15,5))
sns.barplot(x='Promo2',y='Sales',data=rossm,ax= axis1)      #销售额的对比分析
sns.barplot(x='Promo2',y='Customers',data=rossm,ax= axis2)   #用户数的对比分析
```

```python
sales['StateHoliday'].loc[sales['StateHoliday']==0]='0'
sns.countplot(x='StateHoliday',data= sales)                #节假日数据对比分析
fig,(axis1,axis2) = plt.subplots(1,2,figsize=(15,5))
sns.barplot(x='StateHoliday',y='Sales',data=sales,ax= axis1)              #节假日销售额对比分析
sns.barplot(x='StateHoliday',y='Customers',data=sales[(sales['StateHoliday']!=0) & sales['Sales']>0],ax= axis2)  #节假日用户数量对比分析
```

```python
sns.countplot(x='SchoolHoliday',data= sales)                               #学校节假日数量对比
fig,(axis1,axis2) = plt.subplots(1,2,figsize=(15,5))
sns.barplot(x='SchoolHoliday',y='Sales',data=sales,ax= axis1)              #学校节假日平均销售量
sns.barplot(x='SchoolHoliday',y='Customers',data=sales,ax= axis2)           #学校节假日平均用户量分析
```

```python
fig,(axis1,axis2) = plt.subplots(2,1,figsize=(15,10))                  
sns.boxplot(sales['Customers'],ax = axis1)                      #        销售表每日客户箱型图
sales['Customers'].plot(kind='hist',bins=70,xlim=(0,4000),ax=axis2)       #销售表每日客户频率分布图
```

```python
fig,(axis1,axis2) = plt.subplots(2,1,figsize=(15,10))
sns.boxplot(sales['Sales'],ax = axis1)
sales['Sales'].plot(kind='hist',bins=70,xlim=(0,25000),ax=axis2)
```

```python
aver_sales_custom = sales.groupby('Store')['Sales','Customers'].mean()
sales_custom = pd.DataFrame({'Store':aver_sales_custom.index,'Sales':aver_sales_custom['Sales'],
                             'Customers':aver_sales_custom['Customers']},columns=['Store','Sales','Customers'])
store = store.merge(sales_custom,on='Store')
store.head()
```

```python
sns.countplot(x=store['StoreType'],order=['a','b','c','d'])
fig,(axis1,axis2) = plt.subplots(1,2,figsize=(15,5))
sns.barplot(x='StoreType',y='Sales',data=store,ax=axis1)
sns.barplot(x='StoreType',y='Customers',data=store,ax=axis2)
```

```python
sns.countplot(x=store['Assortment'],order=['a','b','c'])
fig,(axis1,axis2) = plt.subplots(1,2,figsize=(15,5))
sns.barplot(x='Assortment',y='Sales',order=['a','b','c'],data=store,ax=axis1)
sns.barplot(x='Assortment',y='Customers',order=['a','b','c'],data=store,ax=axis2)
```

```python
store_id = 6
store_data = sales[sales['Store']==store_id]
average_store_sales = store_data.groupby('Date')["Sales"].mean()
# Get year, and month when Competition started
y = store["CompetitionOpenSinceYear"].loc[store["Store"]  == store_id].values[0]
m = store["CompetitionOpenSinceMonth"].loc[store["Store"] == store_id].values[0]

# Plot 
ax = average_store_sales.plot(legend=True,figsize=(15,4),marker='o')
ax.set_xticks(range(len(average_store_sales)))
ax.set_xticklabels(average_store_sales.index.tolist(), rotation=90)

# Since all data of store sales given in rossmann_df starts with year=2013 till 2015,
# So, we need to check if year>=2013 and y & m aren't NaN values.
if y >= 2013 and y == y and m == m:
    plt.axvline(x=((y-2013) * 12) + (m - 1), linewidth=3, color='grey')
```
